# Bubble Chart

```python
import plotly.express as px

# Aggregating income and expenditure by region
income_cols = ['eainc', 'reg_sal', 'season_sal', 'interest', 'pension', 'dividends', 'net_ret', 'net_mfg', 'net_trans']
exp_cols = ['food', 'housing_water', 'transport', 'education', 'health', 'miscellaneous']

# Summarizing the data by region
df_agg = df_eda.groupby('w_regn').agg({
    **{col: 'sum' for col in income_cols + exp_cols}
}).reset_index()

# Calculate net balance for potential investment analysis
df_agg['total_income'] = df_agg[income_cols].sum(axis=1)
df_agg['total_expenditure'] = df_agg[exp_cols].sum(axis=1)
df_agg['net_balance'] = df_agg['total_income'] - df_agg['total_expenditure']

# Plotting the bubble chart
fig = px.scatter(df_agg, x='total_income', y='total_expenditure', 
                 size='net_balance', color='w_regn',
                 hover_data=['net_balance'],
                 title='High Potential Investment Sectors by Region',
                 labels={'w_regn': 'Region', 'total_income': 'Total Income', 'total_expenditure': 'Total Expenditure'},
                 template='plotly_dark')

fig.update_traces(marker=dict(opacity=0.7, line=dict(width=1, color='white')))
fig.show()
```

## Scatter Plots

```python
# Calculate IQR for Total Income
Q1_income = df_eda['toinc'].quantile(0.25)
Q3_income = df_eda['toinc'].quantile(0.75)
IQR_income = Q3_income - Q1_income

# Calculate IQR for Total Expenditure
Q1_exp = df_eda['totex'].quantile(0.25)
Q3_exp = df_eda['totex'].quantile(0.75)
IQR_exp = Q3_exp - Q1_exp

# Filter out outliers
df_filtered = df_eda[
    (df_eda['toinc'] >= (Q1_income - 1.5 * IQR_income)) & 
    (df_eda['toinc'] <= (Q3_income + 1.5 * IQR_income)) &
    (df_eda['totex'] >= (Q1_exp - 1.5 * IQR_exp)) & 
    (df_eda['totex'] <= (Q3_exp + 1.5 * IQR_exp))
].copy()

# Categorize households into 'Debt' or 'Within Means'
df_filtered['status'] = df_filtered.apply(lambda row: 'Debt' if row['totex'] > row['toinc'] else 'Within Means', axis=1)

# Aesthetic settings
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Scatter plot without outliers
sns.scatterplot(data=df_filtered, x='toinc', y='totex', hue='status', 
                palette={'Debt': '#e74c3c', 'Within Means': '#2ecc71'}, alpha=0.7, s=100)

# Plotting the equality line (spending = income)
plt.plot([df_filtered['toinc'].min(), df_filtered['toinc'].max()], 
         [df_filtered['toinc'].min(), df_filtered['toinc'].max()], 
         color='black', linestyle='--', label='Spending = Income')

# Labels and title
plt.xlabel('Total Income', fontsize=12, labelpad=10)
plt.ylabel('Total Expenditure', fontsize=12, labelpad=10)
plt.title('Income vs. Expenditure: Debt Indicator (Outliers Removed)', fontsize=16, weight='bold')
plt.legend(title='Financial Status')

# Display plot
plt.show()
```

```python
# Calculate Savings (Income - Expenditure)
df_eda['savings'] = df_eda['toinc'] - df_eda['totex']

# Aesthetic settings
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Boxplot of Savings by Income Decile
sns.boxplot(data=df_eda, x='npcinc', y='savings', palette='coolwarm')

# Labels and title
plt.xlabel('Income Decile (National)', fontsize=12, labelpad=10)
plt.ylabel('Savings (Income - Expenditure)', fontsize=12, labelpad=10)
plt.title('Savings by Income Decile', fontsize=16, weight='bold')

# Show plot
plt.show()
```

### Spearman Correlation Coefficient

```python
# Spearman Rank Correlation
def spearman_correlation(df, variables, title="Spearman Rank Correlation Heatmap"):
    """
    Calculates Spearman correlation and visualizes it as a heatmap.
    
    Parameters:
    - df: DataFrame containing the dataset
    - variables: List of column names to include in the correlation matrix
    - title: Title of the heatmap
    """
    # Compute Spearman correlation
    corr = df[variables].corr(method='spearman')

    # Plot the heatmap
    plt.figure(figsize=(14, 10))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=0.5, cbar=True)
    plt.title(title, fontsize=16, weight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.show()

# Select relevant financial and expenditure variables
variables = [
    'toinc', 'totex', 'totdis', 'percapita', 
    'wages', 'cash_abroad', 'cash_domestic', 
    'rentals_rec', 'interest', 'pension', 'dividends',
    'food', 'nfood', 'cloth', 'transport', 'health', 'education'
]

# Apply Spearman correlation and plot
spearman_correlation(df_eda, variables, title="Spearman Rank Correlation: Income and Expenditure")
```

# 7. Data Mining: Association Rule Mining

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Example dataset with income and expenditures
df = pd.DataFrame({
    "Household": ["A", "B", "C", "D", "E"],
    "TOINC": [10000, 20000, 15000, 30000, 25000],  # Total income
    "BREAD": [2500, 4000, 3000, 7500, 6000],       # Bread expenditure
    "MEAT": [2500, 5000, 3500, 8000, 6200],        # Meat expenditure
    "FISH": [1000, 3000, 2000, 5000, 4200],        # Fish expenditure
    "VEG": [2000, 3500, 2500, 7000, 5600]          # Vegetables expenditure
})

# Step 1: Convert expenditures to percentage of income
for col in ["BREAD", "MEAT", "FISH", "VEG"]:
    df[col] = (df[col] / df["TOINC"]) * 100  # Convert to percentage

# Step 2: Define the threshold (e.g., around 25% Â± 5% tolerance)
threshold = 25
tolerance = 5  # Allow variation (e.g., 20% to 30%)

# Step 3: Convert spending to binary (1 if within threshold, 0 otherwise)
for col in ["BREAD", "MEAT", "FISH", "VEG"]:
    df[col] = ((df[col] >= threshold - tolerance) & (df[col] <= threshold + tolerance)).astype(int)

# Drop income column
df_filtered = df.drop(columns=["TOINC"])

# Step 4: Apply Apriori Algorithm
frequent_itemsets = apriori(df_filtered.iloc[:, 1:], min_support=0.6, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Step 5: Display strong associations
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
